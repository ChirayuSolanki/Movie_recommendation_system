```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
#merging two datasets
movies = movies.merge(credits,on = "title")
movies.shape
```

```python
movies.info(1)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.shape
```

```python
# finding missing data
movies.isnull().sum()
```

```python
#dropping missing data
movies.dropna(inplace = True)
```

```python
#finding dupluicate data
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies['cast'][0]
```

```python
import ast
def convert3(obj):
    count = 0
    L = []
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['cast'][0]
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
    return L
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#removing space from keywords,cast,crew,genres
def space(obj):
    L = []
    for i in obj:
        L.append(i.replace(" ",""))
    return L
```

```python
movies['keywords']=movies['keywords'].apply(space)
movies['cast']=movies['cast'].apply(space)
movies['crew']=movies['crew'].apply(space)
movies['genres']=movies['genres'].apply(space)
```

```python
movies.head()
```

```python
# making tags
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
#new dataframe
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head()
```

```python
#converting into string
new['tags'] = new['tags'].apply(lambda x:" ".join(x))
```

```python
new['tags'][0]
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new['tags']=new['tags'].apply(stem)
```

```python
new.head()
```

```python
new['tags'][0]
```

```python
# now converting the tags into vector also known as text vwctorization
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features = 5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
vector
```

```python
cv.get_feature_names()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity =  cosine_similarity(vector)
```

```python
similarity
```

```python
new.head()
```

```python
similarity[119]
```

```python
sorted(list(enumerate(similarity[119])),reverse = True,key = lambda x:x[1])
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distance = sorted(list(enumerate(similarity[index])),reverse = True,key = lambda x:x[1])
    for i in distance[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
